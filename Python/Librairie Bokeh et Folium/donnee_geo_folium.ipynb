```python
# Visualiser des données géographiques avec Folium

**Folium** est une bibliothèque Python ultra-populaire utilisée pour créer des **cartes interactives**.

Si tu as déjà utilisé Google Maps ou OpenStreetMap sur un site web, tu as une idée du résultat : tu peux zoomer, déplacer la carte, et cliquer sur des marqueurs pour voir des informations.

Techniquement, Folium fait le pont entre la puissance de manipulation de données de **Python** (comme Pandas) et la souplesse de **Leaflet.js**, une bibliothèque JavaScript de référence pour la cartographie.

---

### 🌟 Ce qu'on peut faire avec Folium
1.  **Afficher des fonds de cartes** : Tu peux choisir entre plusieurs styles (OpenStreetMap, Stamen Terrain, ou même des vues satellites via des fournisseurs tiers).
2.  **Ajouter des marqueurs** : Placer des points précis avec des icônes, des couleurs et des fenêtres surgissantes (**popups**).
3.  **Créer des cartes de chaleur (Heatmaps)** : Visualiser la densité de certains événements (par exemple, les zones où il y a le plus d'accidents de la route).
4.  **Réaliser des cartes Choroplèthes** : Colorer des régions (pays, départements) en fonction d'une valeur statistique (ex: densité de population par région).

---

### 💻 Un exemple tout simple
Voici comment créer une carte centrée sur Paris avec un marqueur sur la Tour Eiffel en seulement quelques lignes :
```

```python

import folium

# 1. Créer la carte de base centrée sur Paris
carte = folium.Map(location=[48.8566, 2.3522], zoom_start=12)

# 2. Ajouter un marqueur pour la Tour Eiffel
folium.Marker(
    location=[48.8584, 2.2945],
    popup="Tour Eiffel",
    tooltip="Cliquez pour voir",
    icon=folium.Icon(color="red", icon="info-sign")
).add_to(carte)

# 3. Sauvegarder en fichier HTML
carte.save("ma_carte.html")
```


### 🧐 Pourquoi utiliser Folium plutôt qu'un autre outil ?

| Avantages | Inconvénients |
| :--- | :--- |
| **Simple** : Quelques lignes suffisent pour un résultat pro. | **Lourd** : Si tu as 100 000 points, la carte peut devenir lente dans le navigateur. |
| **Interactif** : Zoom, clic, couches superposables. | **Export HTML** : C'est conçu pour le web, moins pour l'impression papier haute définition. |
| **Intégration** : Fonctionne parfaitement avec les DataFrames Pandas. | **JavaScript nécessaire** : Le rendu final nécessite un navigateur pour être affiché. |
