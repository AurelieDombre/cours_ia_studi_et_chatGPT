## Exo 5 : (Résumé des compétences acquises) Mini-système de classement des utilisateurs

Le responsable souhaite que :
* Les prénoms et noms soient affichés avec une majuscule,
* Les utilisateurs soient classés selon leur tranche d’âge,
* Un compteur affiche combien de personnes appartiennent à chaque catégorie.



````python
# code fourni :
utilisateurs = [
       {"prenom": "lucie", "nom": "duval", "age": 12},
       {"prenom": "mathis", "nom": "durand", "age": 17},
       {"prenom": "caroline", "nom": "dubois", "age": 28},
       {"prenom": "robert", "nom": "lefevre", "age": 70}
   ]

# Initialisation des compteurs
enfants = 0
ados = 0
adultes = 0
seniors = 0

# À compléter ici

for utilisateur in utilisateurs:
   prenom = utilisateur["prenom"].capitalize()
   nom = utilisateur["nom"].capitalize()
   age = utilisateur["age"]

   print(f"Nom et prénom : {nom} {prenom}")

   if age < 13:
       enfants += 1
   elif age <= 17:
       ados += 1
   elif age < 64:
       adultes += 1
   else:
       seniors += 1

print("\n--- Résumé ---")
print("Enfants :", enfants)
print("Adolescents :", ados)
print("Adultes :", adultes)
print("Seniors :", seniors)


````
