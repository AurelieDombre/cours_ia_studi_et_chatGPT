## 1. Templates (HTML dynamique)

Créer un dossier `templates/` dans ton app.

Exemple `templates/home.html` :
```html
<!DOCTYPE html>
<html>
<head><title>Accueil</title></head>
<body>
<h1>{{ titre }}</h1>
<ul>
  {% for article in articles %}
    <li>{{ article.titre }}</li>
  {% endfor %}
</ul>
</body>
</html>
```

### Vue avec template
```python
from django.shortcuts import render
from .models import Article

def home(request):
    articles = Article.objects.all()
    return render(request, 'home.html', {'titre': 'Bienvenue', 'articles': articles})
```

## 2. ORM Django (requêtes avancées)

### CRUD
```python
# Create
Article.objects.create(titre='Test', contenu='...')

# Read
Article.objects.all()
Article.objects.filter(titre__icontains='test')

# Update
article = Article.objects.get(id=1)
article.titre = 'New'
article.save()

# Delete
article.delete()
```

### Requêtes avancées
```python
from django.db.models import Count

Article.objects.values('titre').annotate(nb=Count('id'))
Article.objects.order_by('-date_creation')
```

## 3. Django REST Framework (API)

### Installation
```bash
pip install djangorestframework
```

Ajouter dans `INSTALLED_APPS` :
```python
'rest_framework'
```

### Serializer
```python
from rest_framework import serializers
from .models import Article

class ArticleSerializer(serializers.ModelSerializer):
    class Meta:
        model = Article
        fields = '__all__'
```

### Vue API
```python
from rest_framework.decorators import api_view
from rest_framework.response import Response
from .models import Article
from .serializers import ArticleSerializer

@api_view(['GET'])
def api_articles(request):
    articles = Article.objects.all()
    serializer = ArticleSerializer(articles, many=True)
    return Response(serializer.data)
```

### URL API
```python
from django.urls import path
from .views import api_articles

urlpatterns = [
    path('api/articles/', api_articles),
]
```
