# Django : Récapitulatif complet (installation → base de données → données)

Guide pas à pas pour démarrer un projet Django avec PyCharm.

## 1. Installation de Django

```bash
pip install django
```

Vérifier l'installation :

```bash
django-admin --version
```

## 2. Création d’un projet Django

```bash
django-admin startproject mon_projet
cd mon_projet
```

Structure générée :
- `manage.py`
- dossier `mon_projet/` (settings, urls, etc.)

## 3. Lancer le serveur

```bash
python manage.py runserver
```

Accéder à : http://127.0.0.1:8000/

## 4. Création d’une application

```bash
python manage.py startapp mon_app
```

Ajouter l’app dans `settings.py` :

```python
INSTALLED_APPS = [
    ...,
    'mon_app',
]
```

## 5. Création des modèles (BDD)

Dans `models.py` :

```python
from django.db import models

class Article(models.Model):
    titre = models.CharField(max_length=200)
    contenu = models.TextField()
    date_creation = models.DateTimeField(auto_now_add=True)

    def __str__(self):
        return self.titre
```

## 6. Migrations (création de la base)

```bash
python manage.py makemigrations
python manage.py migrate
```

➡️ Django crée automatiquement la base SQLite par défaut.

## 7. Créer un super utilisateur

```bash
python manage.py createsuperuser
```

Accès admin : http://127.0.0.1:8000/admin/

## 8. Enregistrer les modèles dans l’admin

Dans `admin.py` :

```python
from django.contrib import admin
from .models import Article

admin.site.register(Article)
```

## 9. Ajouter des données

### Via l’admin
- Se connecter
- Ajouter des objets

### Via le shell

```bash
python manage.py shell
```

```python
from mon_app.models import Article

Article.objects.create(titre="Test", contenu="Contenu test")
```

## 10. Créer une vue

Dans `views.py` :

```python
from django.http import HttpResponse

def home(request):
    return HttpResponse("Hello Django")
```

## 11. Configurer les URLs

Dans `urls.py` du projet :

```python
from django.contrib import admin
from django.urls import path
from mon_app.views import home

urlpatterns = [
    path('admin/', admin.site.urls),
    path('', home),
]
```

## 12. Utilisation avec PyCharm

- Créer un projet Python
- Configurer un environnement virtuel
- Installer Django
- Utiliser le terminal intégré pour les commandes
- Lancer `manage.py runserver`

PyCharm reconnaît automatiquement la structure Django.
