# Annoter un graphique avec `annotate()`

Ce notebook corrige l'extrait du cours et explique clairement comment annoter un point sur un graphique Matplotlib.

## Erreurs ou points a corriger dans l'extrait du cours de STUDI

- Le code utilise `y3` sans montrer sa definition.
- L'import `math` n'est pas utile ici.
- Le `...` rend l'exemple incomplet.
- Il est preferable de construire un exemple complet et executable.
- Pour `y(x) = 10x`, le point `(6, 60)` est correct, mais il faut bien definir `y = 10 * x`.

````python
import numpy as np
import matplotlib.pyplot as plt

# On cree les valeurs de x.
# Ici, on prend des entiers de 0 a 10 pour lire facilement les points.
x = np.arange(0, 11, 1)

# On definit la fonction du cours : y = 10x.
y = 10 * x

# On cree une figure avec une taille confortable.
plt.figure(figsize=(8, 5))

# On trace la courbe.
plt.plot(x, y, marker="o", color="royalblue")

# On ajoute les informations classiques du graphique.
plt.xlabel("x")
plt.ylabel("y")
plt.title("Exemple : y(x) = 10x")
plt.grid(True)

# On annote le point de coordonnees (6, 60).
# xy      : position exacte du point sur la courbe.
# xytext  : position du texte affiche.
# arrowprops : style de la fleche.
plt.annotate(
    "Point aux coordonnees (6, 60)",
    xy=(6, 60),
    xytext=(2, 80),
    arrowprops={"arrowstyle": "->", "connectionstyle": "arc3,rad=0.3"}
)

# Ajuste automatiquement l'espacement.
plt.tight_layout()

# Affiche le resultat.
plt.show()
````

## Explication de `annotate()`

La fonction `plt.annotate()` sert a ajouter un texte sur le graphique pour attirer l'attention sur une valeur ou un point particulier.

Dans notre exemple :

- `xy=(6, 60)` designe le point de la courbe que l'on veut viser.
- `xytext=(2, 80)` place le texte un peu plus haut et a gauche.
- `arrowprops` permet de dessiner une fleche entre le texte et le point.

## Comprendre `plt.subplot(2, 2, 3)`

L'instruction `plt.subplot(2, 2, 3)` signifie :

- `2` lignes
- `2` colonnes
- `3`e emplacement dans la figure

Matplotlib numerote les positions de gauche a droite, puis de haut en bas :

```text
1  2
3  4
```

Donc :

- `plt.subplot(2, 2, 1)` = en haut a gauche
- `plt.subplot(2, 2, 2)` = en haut a droite
- `plt.subplot(2, 2, 3)` = en bas a gauche
- `plt.subplot(2, 2, 4)` = en bas a droite

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
plt.title("Position 1")

plt.subplot(2, 2, 2)
plt.title("Position 2")

plt.subplot(2, 2, 3)
plt.title("Position 3")

plt.subplot(2, 2, 4)
plt.title("Position 4")

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Donnees
x = np.arange(0, 11, 1)
y1 = 2 * x
y2 = x ** 2
y3 = 10 * x
y4 = np.sqrt(x)

plt.figure(figsize=(10, 8))

# Graphique 1
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("y(x) = 2x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# Graphique 2 : celui du cours
plt.subplot(2, 2, 2)
plt.plot(x, y3, marker="o")
plt.title("y(x) = 10x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.annotate(
    "Point (6, 60)",
    xy=(6, 60),
    xytext=(2, 80),
    arrowprops={"arrowstyle": "->", "connectionstyle": "arc3,rad=0.3"}
)

# Graphique 3
plt.subplot(2, 2, 3)
plt.plot(x, y2)
plt.title("y(x) = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# Graphique 4
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("y(x) = sqrt(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.tight_layout()
plt.show()
````

## A retenir

- `annotate()` sert a ajouter un texte explicatif sur un graphique.
- `xy` indique le point vise.
- `xytext` indique ou afficher le texte.
- `arrowprops` permet de personnaliser la fleche.
- Pour que l'exemple fonctionne, toutes les variables doivent etre definies clairement.
