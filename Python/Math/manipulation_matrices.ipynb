### Exercice 2

Tracez un graphe qui permettra de superposer les recettes quotidiennes effectuées au cours des 2 mois précédents à lʼaide de Matplotlib.

Les résultats du mois antérieur (décembre) sont donnés ici :

[1766, 1543, 8765, 4536, 1100, 4567, 9877, 3467, 2343, 2654, 9273, 1231, 7654, 9876, 6754, 1224, 7680, 7658, 2343, 7653, 7234, 3545, 2343, 3542, 1231, 2354, 1452, 3456, 5436, 4321].

```
import numpy as np
import matplotlib.pyplot as plt

# Tableau NumPy contenant les recettes journalières du mois de décembre
decembre = np.array([1766, 1543, 8765, 4536, 1100, 4567, 9877, 3467, 2343, 2654, 9273, 1231, 7654, 9876, 6754, 1224, 7680, 7658, 2343, 7653, 7234, 3545, 2343, 3542, 1231, 2354, 1452, 3456, 5436, 4321])

# Tableau NumPy contenant les recettes journalières du mois de janvier
janvier = np.array([1100, 1231, 1452, 7680, 3542, 8765, 4567, 4536, 2343, 9273, 1543, 7653, 7654, 2354, 7658, 1231, 4321, 1766, 9877, 9876, 3456, 3545, 2343, 1224, 5436, 2343, 6754, 7234, 2654, 3467])

# Création d'un tableau de temps pour décembre :
# ici on génère les numéros de jours de 1 jusqu'au nombre total de valeurs dans decembre
time1 = np.arange(1, decembre.size + 1, 1)

# Création d'un tableau de temps pour janvier :
# même principe, on génère les jours de 1 à 30
time2 = np.arange(1, janvier.size + 1, 1)

# Trace la courbe de décembre :
# axe des x = jours, axe des y = recettes
# couleur rouge, trait continu
plt.plot(time1, decembre, 'red', linestyle='-')

# Trace la courbe de janvier :
# couleur verte, trait en pointillés
plt.plot(time2, janvier, 'green', linestyle='--')

# Ajoute une légende pour identifier les deux courbes
# Attention : l'ordre de la légende devrait correspondre à l'ordre des courbes tracées
plt.legend(["Décembre", "Janvier"])

# Nom de l'axe horizontal
plt.xlabel("Jours")

# Nom de l'axe vertical
plt.ylabel("Recettes")

# Titre du graphique
plt.title("Recettes du mois de janvier et décembre")

# Affiche le graphique à l'écran
plt.show()
```
