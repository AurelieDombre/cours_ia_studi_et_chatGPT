# 🔎 Le problème de départ
---
Tu compares deux fonctions :

* ( y_1(x) = 2x ) (fonction linéaire)
* ( y_2(x) = e^x ) (fonction exponentielle)

👉 Le souci :
La fonction exponentielle grandit **beaucoup plus vite** que la linéaire.

Par exemple :

* ( e^2 \approx 7 )
* ( e^5 \approx 148 )
* ( e^{10} \approx 22000 )

👉 Résultat : sur un même graphique, la courbe (2x) devient presque invisible (écrasée en bas).

---

# 💡 Solution : les *subplots*

Les **subplots**, c’est simplement :
👉 découper une fenêtre graphique en plusieurs petits graphiques indépendants

Un peu comme ça :

```
+-------------------+-------------------+
|   Graphique 1     |   Graphique 2     |
+-------------------+-------------------+
```

---

# 🧠 Ton code expliqué simplement

```python
fig, axs = plt.subplots(1,2)
```

👉 Tu crées :

* 1 ligne
* 2 colonnes
  ➡️ donc **2 graphiques côte à côte**

👉 `axs` est un tableau :

```python
axs[0]  # premier graphique
axs[1]  # deuxième graphique
```

---

# 📊 Cas concret amélioré (plus lisible)

Voici une version plus propre et pédagogique :

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 50)

y1 = 2 * x
y2 = np.exp(x)

fig, axs = plt.subplots(1, 2)

# Graphique 1 : fonction linéaire
axs[0].plot(x, y1)
axs[0].set_title("y = 2x")

# Graphique 2 : fonction exponentielle
axs[1].plot(x, y2)
axs[1].set_title("y = e^x")

plt.show()
````

---

# 🎯 Ce que tu dois retenir

### 1. `plt.subplots(lignes, colonnes)`

* définit la grille de graphiques

### 2. `axs[i]`

* permet d’accéder à un graphique précis

### 3. Chaque subplot a :

* ses propres axes
* sa propre échelle
* son propre titre

# ⚠️ Petite amélioration importante

y1 = 2 * x
y2 = np.exp(x)

# 🔥 Bonus (super utile)

Tu peux faire une grille 2x2 :

```python
fig, axs = plt.subplots(2, 2)
```

Accès :

```python
axs[0,0]
axs[0,1]
axs[1,0]
axs[1,1]
```

Personnalisons le graphique obtenu à l’exemple précédent. Nous allons ajouter un titre à chaque graphique, activer la grille (« grid ») et ajouter un titre à l’axe des abscisses.

````python
import numpy as np
import matplotlib.pyplot as plt

# 1) Creer les valeurs de x
# np.linspace(debut, fin, nombre_de_points)
# Ici, on fabrique 50 valeurs regulierement espacees entre 1 et 10.
x = np.linspace(1, 10, 50)

# 2) Calculer les valeurs des fonctions
# Bonne pratique : avec NumPy, on evite les boucles quand ce n'est pas necessaire.
# Les operations s'appliquent directement a tout le tableau.
y1 = 2 * x          # calcule y = 2x pour toutes les valeurs de x
y2 = np.exp(x)      # calcule y = e^x pour toutes les valeurs de x

# 3) Creer une figure avec 2 graphiques cote a cote
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# 4) Premier graphique : y = 2x
axs[0].plot(x, y1, color="blue")
axs[0].set_title("y(x) = 2x")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].grid(True)

# 5) Deuxieme graphique : y = e^x
axs[1].plot(x, y2, color="red")
axs[1].set_title("y(x) = exp(x)")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].grid(True)

# 6) Ajuster automatiquement l'espacement entre les graphiques
plt.tight_layout()

# 7) Afficher le resultat
plt.show()
````

Pourquoi cette version est meilleure :

- Elle évite la boucle for, inutile ici.

- Elle utilise np.exp(x) au lieu de math.exp(x) : math.exp() travaille sur une seule valeur, np.exp() travaille directement sur tout un tableau.

- Elle ne crée qu’une seule figure claire avec deux sous-graphiques.

- Elle ajoute ylabel, figsize et tight_layout() pour un affichage plus propre.
En résumé :

    - x est un tableau de 50 valeurs.
    - y1 contient le résultat de 2*x.
    - y2 contient le résultat de e^x.
    - axs[0] correspond au graphique de gauche.
    - axs[1] correspond au graphique de droite.


## Version débutant


````python
# On importe NumPy :
# c'est une bibliotheque tres utile pour manipuler des listes de nombres
# de facon rapide et pratique.
import numpy as np

# On importe pyplot de Matplotlib :
# c'est le module qui permet de tracer des graphiques.
import matplotlib.pyplot as plt


# --------------------------------------------------
# 1) Creation des valeurs de x
# --------------------------------------------------

# np.linspace(1, 10, 50) signifie :
# - commencer a 1
# - finir a 10
# - creer 50 valeurs regulierement espacees entre les deux
#
# Le resultat est un tableau NumPy.
x = np.linspace(1, 10, 50)


# --------------------------------------------------
# 2) Calcul des valeurs de y
# --------------------------------------------------

# Ici, on calcule la fonction y = 2x.
# Comme x est un tableau, NumPy applique le calcul a chaque valeur.
#
# Exemple :
# si x contient [1, 2, 3]
# alors y1 contiendra [2, 4, 6]
y1 = 2 * x

# Ici, on calcule la fonction y = e^x.
# np.exp(x) applique l'exponentielle a chaque valeur du tableau x.
#
# Exemple :
# si x contient [1, 2]
# alors y2 contiendra [e^1, e^2]
y2 = np.exp(x)


# --------------------------------------------------
# 3) Creation de la figure et des sous-graphiques
# --------------------------------------------------

# plt.subplots(1, 2) signifie :
# - 1 ligne de graphiques
# - 2 colonnes de graphiques
#
# La fonction renvoie 2 objets :
# - fig : la figure complete (la grande fenetre)
# - axs : les zones de dessin a l'interieur de la figure
#
# Comme on a demande 2 graphiques, axs contient 2 elements :
# - axs[0] : le graphique de gauche
# - axs[1] : le graphique de droite
fig, axs = plt.subplots(1, 2, figsize=(12, 4))


# --------------------------------------------------
# 4) Premier graphique
# --------------------------------------------------

# On trace la courbe y1 en fonction de x dans le graphique de gauche.
# axs[0].plot(x, y1) veut dire :
# - prendre la zone de dessin numero 0
# - tracer y1 selon x
axs[0].plot(x, y1, color="blue")

# On donne un titre au graphique de gauche.
axs[0].set_title("y(x) = 2x")

# Nom de l'axe horizontal.
axs[0].set_xlabel("x")

# Nom de l'axe vertical.
axs[0].set_ylabel("y")

# On affiche une grille pour mieux lire les valeurs.
axs[0].grid(True)


# --------------------------------------------------
# 5) Deuxieme graphique
# --------------------------------------------------

# Meme principe, mais cette fois pour la fonction y = e^x
# dans le graphique de droite.
axs[1].plot(x, y2, color="red")

axs[1].set_title("y(x) = exp(x)")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].grid(True)


# --------------------------------------------------
# 6) Ajustement de l'affichage
# --------------------------------------------------

# plt.tight_layout() ajuste automatiquement les espacements
# pour eviter que les titres ou les labels se chevauchent.
plt.tight_layout()


# --------------------------------------------------
# 7) Affichage final
# --------------------------------------------------

# plt.show() affiche la figure a l'ecran.
plt.show()
````

Explication des objets importants :
* x C’est un tableau NumPy contenant 50 nombres entre 1 et 10.
* y1 C’est un tableau contenant les valeurs de la fonction y = 2x.
* y2 C’est un tableau contenant les valeurs de la fonction y = e^x.
* fig C’est la figure complète, c’est-à-dire la grande zone qui contient tous les graphiques.
axs C’est un ensemble de sous-graphiques. Ici :
*     axs[0] = graphique de gauche
*     axs[1] = graphique de droite
Pourquoi on n’utilise pas de boucle for :
Avec NumPy, on peut faire des calculs sur tout un tableau d’un seul coup.
* C’est plus court.
* C’est plus lisible.
* C’est généralement plus rapide.
Exemple :
y1 = 2 * x
au lieu de :
for i in range(len(x)):
    y1[i] = 2 * x[i]
Les deux donnent le même résultat, mais la première version est la bonne pratique avec NumPy.

# Schema simple des objets du programme

Voici un schema simple pour visualiser les objets du programme.

## Vue d'ensemble

```text
fig
└── contient 2 sous-graphiques
    ├── axs[0]  -> graphique de gauche
    └── axs[1]  -> graphique de droite
```

Donc :

- `fig` = la grande figure complete
- `axs[0]` = la zone ou on trace `y = 2x`
- `axs[1]` = la zone ou on trace `y = exp(x)`

## Les donnees

```text
x  = [1.0, 1.18, 1.37, 1.55, ..., 10.0]
y1 = [2.0, 2.36, 2.74, 3.10, ..., 20.0]
y2 = [e^1, e^1.18, e^1.37, e^1.55, ..., e^10]
```

En simplifiant :

```text
x[i]   -> une valeur de x
y1[i]  -> la valeur 2 * x[i]
y2[i]  -> la valeur exp(x[i])
```

Cela veut dire que pour chaque position `i` :

- on prend une valeur dans `x`
- on calcule sa valeur associee dans `y1`
- on calcule sa valeur associee dans `y2`

Exemple avec quelques valeurs :

```text
x    = [1, 2, 3]
y1   = [2, 4, 6]
y2   = [e^1, e^2, e^3]
```

## Comment `plot(x, y)` fonctionne

Quand on ecrit :

```python
axs[0].plot(x, y1)
```

Python associe :

- la premiere valeur de `x` avec la premiere valeur de `y1`
- la deuxieme valeur de `x` avec la deuxieme valeur de `y1`
- etc.

Donc on obtient des points comme :

```text
(1, 2)
(2, 4)
(3, 6)
...
```

Et Matplotlib relie ces points pour former la courbe.

Pour l'autre graphique :

```python
axs[1].plot(x, y2)
```

on obtient des points comme :

```text
(1, e^1)
(2, e^2)
(3, e^3)
...
```

## Schema visuel complet

```text
x ----------------------┐
                        ├--> calcul de y1 = 2*x ------> axs[0].plot(x, y1)
                        │
                        └--> calcul de y2 = exp(x) ---> axs[1].plot(x, y2)

fig
├── axs[0] : graphique de y = 2x
└── axs[1] : graphique de y = exp(x)
```

## A retenir

- `x`, `y1`, `y2` sont des tableaux de nombres.
- `fig` est la figure entiere.
- `axs` contient les sous-graphiques.
- `axs[0]` et `axs[1]` sont les zones ou on dessine.
- `plot(x, y)` trace les points `(x[i], y[i])`.
