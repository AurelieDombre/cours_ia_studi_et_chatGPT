# Les vecteur avec numpy

 Un vecteur est une collection ordonnée de nombres. Vous pouvez le visualiser comme une liste ou un tableau à une dimension. Chaque nombre dans le vecteur est un 'élément' ou une 'composante' du vecteur.

Dans l'exemple que vous avez fourni :

- ```import numpy as np``` importe la bibliothèque numpy, qui est très utilisée en Python pour travailler avec des tableaux numériques, y compris des vecteurs.
- ```timeseries = np.linspace(0, 10, 20)``` crée un vecteur appelé timeseries. La fonction ```linspace``` de NumPy génère des nombres espacés uniformément sur un intervalle spécifié.
  - 0 est la valeur de début de l'intervalle.
  - 10 est la valeur de fin de l'intervalle.
  - 20 est le nombre total de points (ou d'éléments) que vous voulez dans votre vecteur, y compris les points de début et de fin.
Le résultat est bien un vecteur de 20 nombres qui commencent à 0 et se terminent à 10, avec des intervalles égaux entre chaque nombre.
Dans de nombreux contextes mathématiques et de programmation, un vecteur est un tableau à une dimension ou une liste de nombres.



## Définir un vecteur de point : (le programme définit les points intermédiaires)

````
import numpy as np

timeseries = np.linspace(0, 10, 20)

print(timeseries)
````

## Définir un vecteur de temps avec ```linspace()``` :

Un vecteur de temps qui s’étend de [0 à 10[ (10 est exclu de l’intervalle), avec un pas de 1

````
import numpy as np

timeseries = np.arange(0, 10, 1)

print(timeseries)
````
