## Kdeplot

La fonction **`kdeplot()`** (Kernel Density Estimation plot) est un outil de visualisation utilisé notamment dans Seaborn pour représenter la **distribution lissée d’une variable**.

---

## 📈 Qu’est-ce que ça fait ?

Un `kdeplot()` transforme des données brutes en une **courbe de densité continue**.

👉 Au lieu de voir des barres (comme un histogramme), tu vois une **courbe fluide** qui estime où les données sont concentrées.

---

## 🔎 Intuition simple

Imagine un histogramme, mais :

* on “arrondit” les barres
* on les relie de manière lisse

👉 Résultat : une courbe qui montre la **probabilité relative** des valeurs.

---

## 📊 À quoi ça ressemble ?

* zones hautes de la courbe → beaucoup de données
* zones basses → peu de données

```text
densité ↑
        /\
       /  \      ← pic = zone fréquente
______/    \______
              → valeurs
```

---

## 🎯 Ce que ça permet de voir

* la **forme de la distribution** (symétrique, biaisée…)
* les **pics** (modes)
* la **dispersion**
* la présence de **multimodalité** (plusieurs pics)

---

## 📌 Cas d’utilisation

### ✔️ Une variable

→ distribution d’un salaire, âge, température…

### ✔️ Deux variables

→ carte de densité (heatmap lissée)

---

## 🌊 Exemple conceptuel

* histogramme → discret, “en marches”
* kdeplot → continu, fluide

---

## ⚠️ Paramètre important

* **bandwidth (bw)** : contrôle le lissage

  * trop petit → courbe très irrégulière
  * trop grand → trop lissé (perte de détails)

---

## 🧠 En résumé

Un **kdeplot** sert à :
👉 estimer la **distribution continue** d’une variable
👉 visualiser la **densité des données de manière fluide**
👉 remplacer ou compléter un histogramme

---

💡 Phrase simple :

> kdeplot = histogramme lissé en une courbe de densité

---

Différence entre KDE et Histogramme :



````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. Génération des données
# -----------------------------
np.random.seed(42)

data = np.random.normal(loc=0, scale=1, size=500)

# -----------------------------
# 2. Histogramme
# -----------------------------
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogramme")
plt.xlabel("Valeurs")
plt.ylabel("Fréquence")

# -----------------------------
# 3. KDE plot
# -----------------------------
plt.subplot(1, 2, 2)
sns.kdeplot(data, fill=True)
plt.title("KDE plot (densité)")
plt.xlabel("Valeurs")
plt.ylabel("Densité")

plt.tight_layout()
plt.show()
````

## Pairplot()

## Boxplot

Un **boxplot** (ou *boîte à moustaches*) est un graphique simple mais très puissant pour résumer une variable quantitative.

### 🔎 Ce qu’il montre

Un boxplot découpe les données en **quartiles** et met en évidence leur dispersion :

* **Minimum** : plus petite valeur (hors valeurs aberrantes)
* **1er quartile (Q1)** : 25 % des données sont en dessous
* **Médiane (Q2)** : valeur centrale (50 % des données de chaque côté)
* **3e quartile (Q3)** : 75 % des données sont en dessous
* **Maximum** : plus grande valeur (hors valeurs aberrantes)

### 📦 Structure du boxplot

* La **boîte** va de Q1 à Q3 → elle contient les 50 % centraux des données
* La **ligne dans la boîte** = la médiane
* Les **moustaches** s’étendent vers les valeurs extrêmes
* Les **points isolés** au-delà des moustaches = valeurs aberrantes (*outliers*)

### ⚠️ Valeurs aberrantes

On considère souvent comme aberrantes les valeurs situées au-delà de :

Q1−1.5×(Q3−Q1) ou Q3+1.5×(Q3−Q1)

### 🎯 Pourquoi c’est utile

* Comparer plusieurs groupes rapidement (ex : salaires par métier)
* Voir la **dispersion** et la **symétrie** des données
* Détecter facilement des **valeurs atypiques**

### 🧠 En résumé

Un boxplot est une **synthèse visuelle compacte** d’une distribution : il montre où se situent les données, comment elles sont réparties, et s’il y a des anomalies.



````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
   'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
   'income': [70000, 80000, 60000, 85000, 90000, 75000, 65000, 95000, 55000, 90000]
})

# Utilisation de seaborn pour créer un boxplot qui montre la distribution du revenu par genre

sns.boxplot(x='gender', y='income', data=data)
plt.title('Distribution du revenu par genre')
plt.show()
````

## Violinplot

Un **violin plot** (ou *diagramme en violon*) est une extension du boxplot qui donne une vision **beaucoup plus détaillée de la distribution** des données.

### 🎻 L’idée générale

Il combine :

* un **boxplot** → pour les statistiques clés (médiane, quartiles)
* une estimation de densité (**KDE**, *Kernel Density Estimation*) → pour voir **la forme de la distribution**

### 🔎 Ce qu’il montre concrètement

* La **largeur du violon** à un niveau donné indique la **densité des données**
  → plus c’est large, plus il y a de valeurs à cet endroit
* La **ligne centrale** correspond souvent à la médiane
* Parfois, on voit aussi les quartiles (comme dans un boxplot)

### 📊 Comment le lire

* Un violon **large au centre** → beaucoup de valeurs autour de la médiane
* Un violon **avec deux bosses** → distribution bimodale (deux groupes de valeurs)
* Un violon **asymétrique** → données non symétriques
* Des extrémités fines → peu de valeurs extrêmes

### 🎯 Pourquoi c’est puissant

Contrairement au boxplot, il permet de :

* voir la **forme complète de la distribution**
* détecter des **structures cachées** (ex : plusieurs pics)
* mieux comprendre la **répartition réelle** des données

### 🧠 En résumé

* **Boxplot** → résumé simple (quartiles + outliers)
* **Violin plot** → résumé + **forme détaillée de la distribution**

👉 On peut dire que le violin plot est un **boxplot enrichi**, idéal quand tu veux aller au-delà d’un simple résumé statistique.


````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
   'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
   'income': [70000, 80000, 60000, 85000, 90000, 75000, 65000, 95000, 55000, 90000]
})

# Utilisation de seaborn pour créer un violin plot qui montre la distribution du revenu par genre
sns.violinplot(x='gender', y='income', data=data)
plt.title('Violin plot du revenu par genre')
plt.show()
````

## Swarmplot

Un **swarmplot** (ou *diagramme en essaim*) est un graphique qui affiche **chaque observation individuellement**, tout en évitant que les points se superposent.

### 🐝 L’idée principale

Contrairement à un boxplot ou un violin plot qui résument les données, le swarmplot montre :
👉 **toutes les valeurs une par une**

Mais avec une astuce :
les points sont **légèrement décalés horizontalement** pour ne pas se chevaucher → ça crée un effet d’“essaim”.

### 🔎 Ce qu’il permet de voir

* La **distribution réelle des données**
* La **densité locale** (zones où les points sont serrés)
* Les **valeurs atypiques** (points isolés)
* La **taille de l’échantillon** (nombre de points visible)

### 📊 Comment le lire

* Zones **denses** → beaucoup de points proches
* Zones **clairsemées** → peu de données
* Points isolés → possibles **outliers**

### 🎯 Pourquoi c’est utile

* Donne une **vision brute et fidèle** des données
* Évite la perte d’information liée aux résumés statistiques
* Très utile pour des **petits à moyens échantillons**

### 🔗 Souvent combiné avec :

* **Boxplot** → pour ajouter médiane et quartiles
* **Violin plot** → pour ajouter la forme de distribution

👉 Exemple : un violin plot + swarmplot permet de voir à la fois
la **forme globale** ET les **points individuels**

### ⚠️ Limites

* Peut devenir illisible avec **beaucoup de données**
* Moins synthétique que les autres graphiques

### 🧠 En résumé

* **Swarmplot** → chaque point compte
* **Boxplot** → résumé statistique
* **Violin plot** → forme de la distribution

C’est le graphique idéal quand tu veux **montrer les données telles qu’elles sont**, sans les cacher derrière des statistiques.


````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
   'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
   'income': [70000, 80000, 60000, 85000, 90000, 75000, 65000, 95000, 55000, 90000]
})

# Utilisation de seaborn pour créer un swarm plot qui montre la distribution du revenu par genre
# avec des points noirs semi-transparents
sns.swarmplot(x='gender', y='income', data=data, color='k', alpha=0.5)
plt.show()
````

# Graphiques multi-panneaux et multi-plots

Seaborn offre d’excellentes fonctionnalités pour créer des visualisations complexes en utilisant des grilles de graphiques, ce qui permet d’analyser des ensembles de données sous plusieurs perspectives à la fois.

## FacetGrid

Un **FacetGrid** est un outil (notamment dans la librairie Seaborn) qui permet de **décomposer une visualisation en plusieurs sous-graphiques** selon des catégories.

### 🧩 L’idée principale

Au lieu de tout afficher sur un seul graphique, on crée une **grille de graphiques** (facettes), où :

* chaque **ligne** ou **colonne** correspond à une catégorie (ex : genre, pays, année…)
* chaque case montre le **même type de graphique**, mais pour un sous-ensemble de données

### 🔎 Ce que ça permet

* Comparer visuellement des distributions ou relations
* Observer l’effet d’une variable catégorielle sur une autre
* Explorer des données complexes de manière structurée

### 📊 Exemple simple

Imaginons :

* X = âge
* Y = salaire
* Catégorie = secteur d’activité

👉 Avec un FacetGrid :

* chaque colonne = un secteur
* chaque graphique = relation âge vs salaire dans ce secteur

### 🎯 Pourquoi c’est puissant

* Permet de **voir des patterns cachés**
* Facilite les **comparaisons multiples**
* Évite de surcharger un seul graphique

### ⚙️ Ce qu’on peut contrôler

* Nombre de lignes/colonnes
* Variable de séparation (*row*, *col*, *hue*)
* Type de graphique dans chaque case (histogramme, scatterplot, etc.)

### 🧠 En résumé

Un **FacetGrid = une grille de visualisations conditionnelles**
👉 idéal pour explorer comment une relation change selon différentes catégories

C’est un outil clé en analyse exploratoire quand tu travailles avec plusieurs variables en même temps.



````python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")

plt.show()
````

## PairGrid

Un **PairGrid** est une version plus flexible de la visualisation des relations entre variables dans un dataset, proposée notamment par Seaborn.

---

## 🧩 L’idée générale

Un **PairGrid** construit une **grille de graphiques** où :

* chaque **ligne** = une variable
* chaque **colonne** = une variable
* chaque case = une relation entre deux variables

👉 On obtient donc une matrice de visualisations.

---

## 📊 À quoi ça ressemble (schéma)

Imaginons 3 variables : A, B, C

```
        A          B          C
A     [diag]    scatter    scatter
B    scatter    [diag]     scatter
C    scatter    scatter    [diag]
```

---

## 🎯 Ce qui le rend puissant

Contrairement à un simple `pairplot`, le PairGrid permet de personnaliser :

### 🔵 Diagonale

* histogrammes
* KDE
* densité

### 🟢 Triangle inférieur

* scatterplots
* régressions

### 🔴 Triangle supérieur

* heatmaps
* corrélations
* autres types de graphes

👉 Chaque zone peut avoir **un type de graphique différent**

---

## 🔎 Exemple concret

Dataset avec :

* âge
* salaire
* expérience

Tu peux faire :

* diagonale → distribution de chaque variable
* triangle inférieur → régressions (âge vs salaire, etc.)
* triangle supérieur → corrélations ou KDE

---

## ⚖️ PairGrid vs Pairplot

| Feature                   | Pairplot           | PairGrid        |
| ------------------------- | ------------------ | --------------- |
| Facilité                  | Très simple        | Plus complexe   |
| Personnalisation          | limitée            | très élevée     |
| Types de plots différents | non                | oui             |
| Usage                     | rapide exploration | analyse avancée |

---

## 🧠 En résumé

Un **PairGrid**, c’est :
👉 une **matrice de relations entre variables**
👉 entièrement **personnalisable selon les zones**
👉 idéal pour une analyse exploratoire poussée

---

💡 En une phrase :

> PairGrid = pairplot, mais avec contrôle total sur chaque partie de la grille

---



````python
import seaborn as sns
import matplotlib.pyplot as plt

# Supposons que 'tips' soit un DataFrame qui vient de seaborn
tips = sns.load_dataset('tips')

# Création d'un PairGrid
g = sns.PairGrid(tips)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, color="C0")  # Utilise 'color' au lieu de 'colors'
g.map_diag(sns.histplot, kde=True, color='k')  # Active le KDE et définit la couleur

plt.show()
````
