# La librairie seaborn

Pour exemple, on utilise un dataset de Seaborn, qui représente au niveau statistique, les chances de survie des passagers du Titanic

```python
import matplotlib.pyplot as plt
import seaborn as sns


# On charge un dataset d'exemple fourni par Seaborn.
# "titanic" contient des informations sur les passagers du Titanic
# (âge, sexe, classe, survie, etc.).
titanic = sns.load_dataset("titanic")
# Le dataset Titanic contient des NaN dans la colonne age. Ça peut casser le modèle logistique.
titanic = titanic.dropna(subset=["age", "survived"])

# head() affiche les 5 premières lignes du tableau.
# C'est pratique pour voir rapidement à quoi ressemblent les données.
print(titanic.head())

# On définit des "tranches" d'âge.
# Ici, Seaborn va regrouper les âges autour de ces bornes
# pour rendre la visualisation plus lisible.
age_bins = [15, 30, 45, 60]

# Version basique
# fig = sns.regplot(
#     data=titanic,
#     x="age",
#     y="survived",
#     logistic=True
# )

# Version moderne
fig = sns.lmplot(
    data=titanic,
    x="age",
    y="survived",
    hue="sex",
    x_bins=age_bins,
    logistic=True,
    height=6,
    aspect=1.2
)

# On fixe les limites de l'axe des x entre 0 et 80 ans
# pour éviter que le graphique ne s'étende trop loin.
fig.set(xlim=(0, 80))

# Affiche le graphique final.
plt.show()
```


lmplot() sert à tracer une relation entre deux variables.
Ici :
- x="age"  : l'âge sur l'axe horizontal
- y="survived" : la survie sur l'axe vertical

La variable "survived" vaut 0 ou 1 :
- 0 = la personne n'a pas survécu
- 1 = la personne a survécu
hue="sex" demande de séparer l'affichage par sexe
avec une couleur différente pour chaque groupe.

x_bins=age_bins regroupe les âges en classes pour simplifier
la lecture de la tendance.

logistic=True est très important ici :
comme "survived" est une variable binaire (0 ou 1),
Seaborn utilise une régression logistique au lieu d'une droite classique.
Cela permet d'estimer une probabilité de survie selon l'âge.

Pour bien comprendre ce graphique, retiens surtout ceci :
- age est la variable explicative.
- survived est la variable à expliquer.
- hue="sex" compare hommes et femmes.
- logistic=True transforme la courbe en estimation de probabilité de survie.
Lecture du résultat :
    - Si la courbe monte, cela suggère une probabilité de survie plus élevée.
    - Si elle descend, cela suggère une probabilité plus faible.
    - Comme il y a une couleur par sexe, tu peux comparer les tendances hommes/femmes selon l’âge.
    - Les traits verticaux représentent le max et le min puis le point au milieu la moyenne du taux de survie
