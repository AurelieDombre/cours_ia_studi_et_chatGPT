## 🧩 1. Décorateur de classe

Un décorateur de classe est une fonction qui prend une classe en entrée et retourne une **nouvelle classe modifiée**.

⚠️ Attention : cela peut rendre le code difficile à lire si mal utilisé.


```

def decorate_class(cls):
    class ClassWrapper(cls):
        def MyMethod(self):
            return "My decorated method implementation"
    return ClassWrapper

@decorate_class
class InitialClass:
    def MyMethod(self):
        return "My method implementation"

obj = InitialClass()
print(obj.MyMethod())
```


### 🔍 Explication

- `decorate_class` reçoit une classe `cls`
- Il crée une **nouvelle classe** `ClassWrapper` qui hérite de `cls`
- Cette nouvelle classe **redéfinit** la méthode `MyMethod`
- Le décorateur remplace `InitialClass` par `ClassWrapper`

👉 Résultat :
Même si `InitialClass` avait une méthode, elle est **écrasée** par celle du décorateur.



## ⚠️ 2. Bonnes pratiques et pièges

### Ordre des décorateurs
- Ils s'appliquent **de bas en haut**
- Et **de droite à gauche**

### Perte des métadonnées
Un décorateur peut faire perdre :
- le nom de la fonction (`__name__`)
- sa documentation (`__doc__`)

👉 Solution : utiliser `functools.wraps`


```

from functools import wraps

def decorate(func):
    @wraps(func)
    def wrapper():
        print("before")
        func()
        print("after")
    return wrapper

@decorate
def Hello():
    """my method documentation"""
    print("Hello!")

print(Hello.__name__)
print(Hello.__doc__)
Hello()
```


### 🔍 Explication

- `@wraps(func)` copie les métadonnées de `func` vers `wrapper`
- Sans cela :
  - `Hello.__name__` deviendrait `"wrapper"`
  - la docstring serait perdue

👉 C'est une **bonne pratique essentielle**



## ✅ Résumé

- Un décorateur modifie une fonction ou une classe
- Les décorateurs de classe remplacent la classe originale
- Attention à la lisibilité du code
- Toujours utiliser `@wraps` pour les fonctions décorées
