## ð§Đ 1. DÃĐcorateur de classe

Un dÃĐcorateur de classe est une fonction qui prend une classe en entrÃĐe et retourne une **nouvelle classe modifiÃĐe**.

â ïļ Attention : cela peut rendre le code difficile Ã  lire si mal utilisÃĐ.


```

def decorate_class(cls):
    class ClassWrapper(cls):
        def MyMethod(self):
            return "My decorated method implementation"
    return ClassWrapper

@decorate_class
class InitialClass:
    def MyMethod(self):
        return "My method implementation"

obj = InitialClass()
print(obj.MyMethod())
```


### ð Explication

- `decorate_class` reÃ§oit une classe `cls`
- Il crÃĐe une **nouvelle classe** `ClassWrapper` qui hÃĐrite de `cls`
- Cette nouvelle classe **redÃĐfinit** la mÃĐthode `MyMethod`
- Le dÃĐcorateur remplace `InitialClass` par `ClassWrapper`

ð RÃĐsultat :
MÃŠme si `InitialClass` avait une mÃĐthode, elle est **ÃĐcrasÃĐe** par celle du dÃĐcorateur.

### Perte des mÃĐtadonnÃĐes
Un dÃĐcorateur peut faire perdre :
- le nom de la fonction (`__name__`)
- sa documentation (`__doc__`)

ð Solution : utiliser `functools.wraps`


```

from functools import wraps

def decorate(func):
    @wraps(func)
    def wrapper():
        print("before")
        func()
        print("after")
    return wrapper

@decorate
def Hello():
    """my method documentation"""
    print("Hello!")

print(Hello.__name__)
print(Hello.__doc__)
Hello()
```


### ð Explication

- `@wraps(func)` copie les mÃĐtadonnÃĐes de `func` vers `wrapper`
- Sans cela :
  - `Hello.__name__` deviendrait `"wrapper"`
  - la docstring serait perdue

ð C'est une **bonne pratique essentielle**
