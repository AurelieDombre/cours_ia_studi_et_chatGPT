## 🔹 Exemple 1 : Décorateur simple

On modifie le résultat d’une fonction.

````python
def decorate(func):
    def wrapper(*args):
        y = func(*args)
        result = y * 100
        print(result)
        return result
    return wrapper

@decorate
def divide(a, b):
    return a / b

divide(10, 100)
````

👉 Ici, la fonction `divide` est remplacée par `wrapper`.

Donc :
```python
divide = decorate(divide)
```

## 🔹 Exemple 2 : Décorateur avec paramètres

On ajoute des messages avant et après.

````python
def decorate_with_params(message_before, message_after):
    def decorate(func):
        def wrapper():
            print(message_before)
            func()
            print(message_after)
        return wrapper
    return decorate

@decorate_with_params("Before", "After")
def hello():
    print("Hello!")

hello()
````

👉 Il y a 3 niveaux :
1. Fonction avec paramètres
2. Décorateur
3. Wrapper

## 🔹 Exemple 3 : Empilement de décorateurs

````python
def decorate_outer(func):
    def wrapper():
        print("begin outer decorate")
        func()
        print("end outer decorate")
    return wrapper

def decorate_inner(func):
    def wrapper():
        print("begin inner decorate")
        func()
        print("end inner decorate")
    return wrapper

@decorate_outer
@decorate_inner
def my_fun():
    print("Initial function")

my_fun()
````

👉 Équivalent à :
```python
my_fun = decorate_outer(decorate_inner(my_fun))
```

Ordre d’exécution :
- outer commence
- inner
- fonction
- inner termine
- outer termine

## 🎯 Exercice

Créer un décorateur qui affiche :
- "Début"
- exécute la fonction
- "Fin"

````python
def decorate(func):
    def wrapper():
        print("Début")
        func()
        print("Fin")
    return wrapper

@decorate
def my_func():
    print("Executer la fonction")

my_func()

````
