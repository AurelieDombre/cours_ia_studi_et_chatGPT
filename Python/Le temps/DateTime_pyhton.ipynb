# ⏰ Le temps avec Python

# Créer un objet de type datetime

````python
import datetime

currentTime = datetime.datetime.now()
print(currentTime)

print(currentTime.day)
print(currentTime.year)
print(currentTime.hour)
print(currentTime.minute)
print(currentTime.second)
````

Une des méthodes couramment utilisée avec les objets datetime est ```strftime()```.
Cette fonction permet d’extraire une information spécifique de notre objet datetime pour pouvoir l’afficher sous forme d’une chaine de caractère.

````python
import datetime

time1 = datetime.datetime(2025, 1, 1)
print(time1)

month = time1.strftime("%B")
print(month)
````

On voit que le fait d’avoir spécifié l’argument ‘%B’ dans l’appel à la fonction strftime nous renvoie le mois correspondant à l’objet datetime, écrit en toutes lettres.

La fonction **strftime** permet d’extraire plusieurs informations d’un objet datetime. Les différents arguments utilisables avec cette fonction sont repris dans le tableau suivant :

| Argument | Description | Exemple |
|---|---|---|
| `%a` | Jour de la semaine en abrégé | `Mon` |
| `%A` | Jour de la semaine complet | `Monday` |
| `%w` | Jour de la semaine (`0` à `6`) | `1` |
| `%d` | Jour du mois (`1` à `31`) | `10` |
| `%b` | Nom du mois en abrégé | `Jan` |
| `%B` | Nom complet du mois | `January` |
| `%m` | Mois de l’année (`1` à `12`) | `1` |
| `%y` | Année (version courte) | `24` |
| `%Y` | Année (version complète) | `2024` |
| `%H` | Heure (`0` à `23`) | `12` |
| `%I` | Heure (`0` à `12`) | `04` |
| `%p` | AM/PM | `PM` |
| `%M` | Minute (`0` à `59`) | `10` |
| `%S` | Seconde (`0` à `59`) | `15` |
| `%f` | Microseconde (`0` à `999999`) | `15200` |
| `%z` | Décalage par rapport au temps UTC | `+0100` |
| `%Z` | Zone géographique | `CST` |
| `%j` | Jour de l’année (`1` à `366`) |  |
| `%U` | Semaine de l’année (dimanche = 1er jour de la semaine) (`0` à `53`) | `15` |
| `%W` | Semaine de l’année (lundi = 1er jour de la semaine) (`0` à `53`) | `12` |
| `%c` | Temps complet au format local | `Fri Oct 04 18:24:04 2024` |
| `%C` | Siècle courant | `21` |
| `%x` | Moment de l’année au format local | `04/10/2024` |
| `%X` | Temps au format local | `18:24:04` |

Le temps « UTC » signifie « Universal Time Coordinated ». C’est un temps universel et standardisé. En fonction de la zone géographique, on définit des zones horaires spécifiques.
